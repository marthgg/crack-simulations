```python
filenames = ['/home/users/marthgg/elastic_properties/t_300/deformX/log.lammps',
             '/home/users/marthgg/elastic_properties/t_300/deformY/log.lammps',
             '/home/users/marthgg/elastic_properties/t_300/deformZ/log.lammps']

deform = ['x', 'y', 'z']

k = 0

for filename in filenames:
    
    log = lammps_logfile.File(filename)

    time = log.get('Time')

    lx = log.get('v_lx')
    ly = log.get('v_ly')
    lz = log.get('v_lz')

    lx_start = np.mean(lx[200:400])
    ly_start = np.mean(ly[200:400])
    lz_start = np.mean(lz[200:400])

    lx_end = lx[1000:]
    ly_end = ly[1000:]
    lz_end = lz[1000:]

    strain_x = (lx_start - lx_end)/lx_start
    strain_y = (ly_start - ly_end)/ly_start
    strain_z = (lz_start - lz_end)/lz_start
    
    strain_xy = (((ly_start - ly_end)/lx_start) + ((lx_start - lx_end)/ly_start))/2

    pxx = log.get('Pxx')
    pyy = log.get('Pyy')
    pzz = log.get('Pzz')
    pxy = log.get('Pxy')
    pxz = log.get('Pxz')
    pyz = log.get('Pyz')
    
    plt.figure(1)
    
    if deform[k] == 'x':
        plt.plot(strain_x, pxx[1000:])
        
        [c_11, b] = np.polyfit(strain_x, pxx[1000:], 1)
        plt.plot(strain_x, c_11*strain_x + b, 'r', label = 'C_11 = %f' %(c_11/10**5))

    if deform[k] == 'y':
        plt.plot(strain_y, pxx[1000:])
        
        [c_12, b] = np.polyfit(strain_y, pxx[1000:], 1)
        plt.plot(strain_y, c_12*strain_y + b, 'c', label = 'C_12 = %f' %(c_12/10**5))
        
    if deform[k] == 'z':
        plt.plot(strain_z, pzz[1000:])

        [c_33, b] = np.polyfit(strain_z, pzz[1000:], 1)
        plt.plot(strain_z, c_33*strain_z + b, 'b', label = 'C_33 = %f' %(c_33/10**5))
        
        plt.plot(strain_z, pxx[1000:])
        
        [c_13, b] = np.polyfit(strain_z, pxx[1000:], 1)
        plt.plot(strain_z, c_13*strain_z + b, 'y', label = 'C_13 = %f' %(c_13/10**5))
    
    plt.xlabel('strain')
    plt.ylabel('stress')
    plt.legend()
    
    k += 1

    
poissons_ratio = (1+(c_11/c_12))**-1
print('Poissons ration: ', poissons_ratio)

E = (c_11 + c_12)/((1+poissons_ratio)*(1-(2*poissons_ratio)))
print('E: ', E/10**5)

```
