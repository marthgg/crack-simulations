```python
import lammps_logfile
from regex_file_collector import Collector
from regex_file_collector.utils import floating_number_pattern
import matplotlib.pyplot as plt
```

```python
# Rotation: 30 degrees
paths = ['/home/users/marthgg/2021_05_simulations_quartz_relaxed_structure/angle_30/relaxation2/log.lammps']

# Rotation: 60 degrees
#paths = ['/home/users/marthgg/2021_05_simulations_quartz_relaxed_structure/angle_60/relaxation2/log.lammps']

# No rotation
#paths = ['/home/users/marthgg/2021_05_simulations_quartz_relaxed_structure/x3000_y200/relaxation2/log.lammps']

for path in paths:
    filename = path
    log = lammps_logfile.File(filename)

    pzz = log.get('Pzz')
    pxx = log.get('Pxx')
    pyy = log.get('Pyy')

    temp = log.get('Temp')

    plt.figure(1)
    plt.plot(pzz/10, label='pzz')
    plt.plot(pxx/10, label='pxx')
    plt.plot(pyy/10, label='pyy')
    plt.ylabel('Pressure (MPa)')
    plt.legend()

    plt.figure(2)
    plt.plot(temp)
    plt.ylabel('Temperature')
```
